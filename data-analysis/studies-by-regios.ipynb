```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('../ctg-geo-filtered.csv')
df.head()
```

```python
fig = px.density_map(df, lat='latitude', lon='longitude', radius=10,
                        center=dict(lat=20, lon=20), zoom=1,
                        map_style="open-street-map")
fig.show()
```
